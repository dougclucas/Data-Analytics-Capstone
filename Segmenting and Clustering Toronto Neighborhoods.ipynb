```python
Table = pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
```

Pulling the tables from the website in using pandas.  Selecting the first table below.

```python
df = Table[0]
```

Eliminating "Not assigned" boroughs from the data frame

```python
DF = df[df.Borough != 'Not assigned']
DF
```

```python
DF.shape
```

```python
##Section 2: Joining the latitude and longitude coordinates with the df
```

```python
latlong = pd.read_csv('http://cocl.us/Geospatial_data')
```

```python
Table = pd.merge(DF, latlong, on = 'Postal Code')
Table
```

```python
## Section 3: Clustering Neighborhoods
```

```python
pip install geopy
```

```python
import numpy as np
import folium
from sklearn.cluster import KMeans
import matplotlib.cm as cm
import matplotlib.colors as colors
```

```python
##Create map of Toronto
Lat = 43.7
Long = -79.4
toronto = folium.Map(location=[Lat, Long], zoom_start=10)
```

```python
#Adding neighborhood markers to the map:

for lat, lng, borough, neighborhood in zip(Table['Latitude'], Table['Longitude'], Table['Borough'], Table['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(toronto) 
```

```python
toronto
```

```python
#Creating a cluster that only includes "Downtown" in Borough Name

dwntwn = Table[Table['Borough'] == 'Downtown Toronto'].reset_index(drop=True)
dwntwn.head()
```

```python
#creating new Downtown map

Lat = 43.65
Long = -79.375
Downtown = folium.Map(location=[Lat, Long], zoom_start=12)

for lat, lng, borough, neighborhood in zip(dwntwn['Latitude'], dwntwn['Longitude'], dwntwn['Borough'], dwntwn['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(Downtown)
```
